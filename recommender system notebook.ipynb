```python
import ast
```

```python
def extract_genre(text):
    A = []
    for i in ast.literal_eval(text):
        A.append(i['name'])
    return A
```

```python
df['genres'] = df['genres'].apply(extract_genre)
```

```python
def extract_top5_actors_names(text):
    A = []
    x = 0
    for i in ast.literal_eval(text):
        if x != 5:
            A.append(i['name'])
            x = x + 1
        else:
            break
    return A
```

```python
df['cast'] = df['cast'].apply(extract_top5_actors_names)
```

```python
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['overview'] = df['overview'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors[0]
```

```python
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:8]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight')
```
